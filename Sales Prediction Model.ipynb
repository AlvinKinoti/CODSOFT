### Visualizing the dataset

##### Finding the highest sales volume recorded

```python
plt.figure(figsize = (5,5))
plt.hist(df["Sales"], bins = 25)
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.grid(True)
plt.title("Frequency of Sales Revenue")
```

##### Comparision between TV and Radio input and Sales

```python
Sales = df["Sales"]
TV = df["TV"]
Radio = df["Radio"]

plt.figure(figsize = (10,10))
plt.scatter(Sales, TV, s = 200)
plt.scatter(Sales, Radio, s = 500, marker = "3")
plt.grid(True)
plt.legend(["TV", "Radio"])
plt.title("Sales vs TV and Radio")
plt.xlabel("Sales")
plt.ylabel("TV and Radio")
```

### Developing the prediction model

```python
X = df.drop(columns = ["Sales"])
y = df["Sales"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, y_train, X_test, y_test = train_test_split(X,y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,y)
```

```python
pred = lr.predict(X)
pred
```

```python
r2_score(pred, y)
```

##### Conclusion : Sales is highly dependent on the input of TV, Radio and Newspaper outlet inputs

### Plotting the results

```python
plt.figure(figsize=(5,5))
plt.scatter(pred, y, s= 250, color = "green", marker = "4")
plt.xlabel("Predicted Values")
plt.ylabel("Sales")
plt.grid(True)
plt.title(" Estimated Values vs Sales")
```
